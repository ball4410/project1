```python
# Dependencies
import requests
import json
import pandas as pd
import gmaps
import numpy as np
```

```python
# Target city
target_city = "San Francisco"

# # Build the endpoint URL
# target_url = ('https://maps.googleapis.com/maps/api/geocode/json?'
#     'address={0}&key={1}').format(target_city, gkey)

# URL for GET requests to retrieve beer data
url = ('https://api.openbrewerydb.org/breweries?by_city={0}').format(target_city)
```

```python
# Print the response object to the console
print(requests.get(url))
```

```python
# Retrieving data and converting it into JSON
print(requests.get(url).json())
```

```python
# Pretty Print the output of the JSON
response = requests.get(url).json()
print(json.dumps(response, indent=4, sort_keys=True))
```

```python
#How many breweries do we have?
len(response)
```

```python
#get lat and long
lat = response[0]["latitude"]
lng = response[0]["longitude"]

# Print the latitude and longitude
print('''
    City: {0}
    Latitude: {1}
    Longitude: {2}
    '''.format(target_city, lat, lng))
```

```python
# Store all the brewery names from the response
types_df = pd.read_csv("./mergedAll.csv")
types_df.head()

types = types_df["Type of Brewery"].unique()

brewery_types_df = pd.DataFrame({"Brewery Type": types})

brewery_types_df['brewery'] = ""
brewery_types_df['address'] = ""
brewery_types_df['website'] = ""
brewery_types_df['latitude'] = ""
brewery_types_df['longitude'] = ""

brewery_types_df
```

```python
for index, row in brewery_types_df.iterrows():
    
    #get brewery type from df
    brew_type = row["Brewery Type"]
    
    print(f"Retrieving results for Index {index}: {brew_type}")
    url = ('https://api.openbrewerydb.org/breweries?by_city={0}&by_type={1}').format(target_city, brew_type)
    response = requests.get(url).json()
    
    try:
        brewery_types_df.loc[index, 'brewery'] = response[0]['name']
        brewery_types_df.loc[index, 'address'] = response[0]['street']
        brewery_types_df.loc[index, 'website'] = response[0]['website_url']
        brewery_types_df.loc[index, 'latitude'] = response[0]['latitude']
        brewery_types_df.loc[index, 'longitude'] = response[0]['longitude']
    except (KeyError, IndexError):
        print("Missing field/result...skipping")

brewery_types_df.to_csv("./SFAPIPUll.csv", index=False, header=True)
brewery_types_df
```

```python
# Target city
locations = brewery_types_df[["latitude", "longitude"]].astype(float)

gmaps.configure(api_key=)

fig = gmaps.figure()
markers = gmaps.marker_layer(locations)

fig.add_layer(markers)
fig

# # Build the endpoint URL
# target_url = ('https://maps.googleapis.com/maps/api/geocode/json?'
#     'address={0}&key={1}').format(target_city, gkey)
```
